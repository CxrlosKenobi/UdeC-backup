```python
# Problema 1
# Programa que calcula promedio de notas según ponderaciones.

def promedioCertamenes(c1, c2, c3, c4, c5):
    ## Promedio de certámenes
    # 10% C1
    # 20% C2
    # 20% C3
    # 20% C4
    # 30% C5
    return (c1 * 0.1 + c2 * 0.2 + c3 * 0.2 + c4 * 0.2 + c5 * 0.3)

def promedioFinal(certamen, tarea, autoevaluacion, profesor):
    ## Promedio final
    # 50% Certámenes
    # 40% Tarea
    # 4% Autoevaluación
    # 6% Profesor
    return (certamen * 0.5 + tarea * 0.4 + autoevaluacion * 0.04 + profesor * 0.06)

def main():
    c1 = float(input("Ingrese nota C1: "))
    c2 = float(input("Ingrese nota C2: "))
    c3 = float(input("Ingrese nota C3: "))
    c4 = float(input("Ingrese nota C4: "))
    c5 = float(input("Ingrese nota C5: "))
    tarea = float(input("Ingrese nota Tarea semestral: "))
    autoevaluacion = float(input("Ingrese nota Autoevaluación: "))
    profesor = float(input("Ingrese nota Desempeño evaluado por profesor: "))

    promedio_1 = round(promedioCertamenes(c1, c2, c3, c4, c5), 3)
    promedio_2 = round(promedioFinal(promedio_1, tarea, autoevaluacion, profesor), 3)

    print(f'\n{promedio_1},{promedio_2}')


if __name__ == '__main__':
    main()
```

```python
# Problema 2
# Extensión del Problema 1, incluyendo estado de aprobación mediante NCR.

def promedioCertamenes():
    c1 = float(input("Ingrese nota C1: "))
    c2 = float(input("Ingrese nota C2: "))
    c3 = float(input("Ingrese nota C3: "))
    c4 = float(input("Ingrese nota C4: "))
    c5 = float(input("Ingrese nota C5: "))
    return round(c1 * 0.1 + c2 * 0.2 + c3 * 0.2 + c4 * 0.2 + c5 * 0.3, 3)

def promedioFinal(certamenes):
    tarea = float(input("Ingrese nota Tarea semestral: "))
    autoevaluacion = float(input("Ingrese nota Autoevaluación: "))
    profesor = float(input("Ingrese nota Desempeño evaluado por profesor: "))

    promedio = (certamenes * 0.5 + tarea * 0.4 + autoevaluacion * 0.04 + profesor * 0.06)

    if tarea >= 4.0:
        # Si la nota de la tarea semestral es mayor o igual a 4.0, el estado es Aprobado
        return round(promedio, 3), "aprobado"
    else:
        # Si no, el estado es Reprobado por NCR
        return "NCR", "reprobado"

def main():
    promedio_1 = promedioCertamenes()

    promedio_2_raw = promedioFinal(promedio_1)
    promedio_2 = promedio_2_raw[0]
    estado = promedio_2_raw[1]

    print(f'\n{promedio_1},{promedio_2}, {estado}')


if __name__ == '__main__':
    main()
```

```python
# Problema 3
# Programa que realiza el análisis por tramo de ingresos del enunciado.

def reajuste(n):
    if n >= 1000000:
        # Mantiene el sueldo original
        return n
    if 500000 < n < 1000000:
        # Aumenta el sueldo en un 5%
        n *= 1.05
    if n < 500000:
        # Aumenta el sueldo en un 10%
        n *= 1.1
    return int(n)

def main():
    n = int(input("Sueldo: "))

    if n > 0 and type(n) == int:
        return print(f'\n{reajuste(n)}')
    else:
        return 1

if __name__ == "__main__":
    main()
```

```python
# Problema 4
# Programar la operación multiplicación (x * y) usando sólo sumas.

def multiplicar(x, y) -> int:
    producto = 0
    while y != 0:
        y -= 1
        producto += x
    
    # Retornamos la sumatoria de "x" en si mismo, "y" veces.
    return producto 

def main():
    x = int(input("x: "))
    y = int(input("y: "))

    if (x >= 0 and type(x) == int) and (y >= 0 and type(y) == int):
        print(f'\n{multiplicar(x, y)}')
    else:
        return 1


if __name__ == "__main__":
    main()    
```

```python
# Problema 5
# Programar la operación potencia (x a la y-ésima potencia) usando sólo sumas.

# Función auxiliar
def aux(x, y) -> int: # Señalamos que la función retorna un int
    # Si y no es 0, podemos hacer la operación de
    # sumar x al resultado de multiplicar x por n - 1

    # Usando recursión, retornamos el resultado de la operación
    return x + aux(x, y - 1) if (y != 0) else 0 

def potencia(x, y) -> int: # Señalamos que la función retorna un int
    # Retornamos operaciones simples por reglas de potencia
    if x == 0:
        return 0
    if y == 0:
        return 1
    if y == 1:
        return x
    
    # Calculamos el resultado de x a la y-ésima potencia
    # Multiplicando x por si mismo n - 1 veces
    return aux(x, potencia(x, y - 1))

def main():
    x = int(input("x: "))
    y = int(input("y: "))

    if (x >= 0 and type(x) == int) and (y >= 0 and type(y) == int):
        return print(f'\n{potencia(x, y)}')
    else:
        return 1

if __name__ == "__main__":
    main()
```
