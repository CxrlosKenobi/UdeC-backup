```python
## 1.
def validar(min,max,mensaje):
    v = int(input(mensaje))
    while v<=min or v>max:
        print("Error, ingrese numero valido.")
        v = int(input(mensaje))
    return v
def media(lista):
    suma = 0
    for i in lista:
        suma = suma + i
    
    cantidad = len(lista)
    promedio = suma/cantidad
    return promedio

def desv(media,lista):
    s = 0
    for i in lista:
        s = s + (i-media)**2
    #terminada la sumatoria, dividimos por el total de datos
    cantidad = len(lista)
    s = s/cantidad 
    #ahora falta sacarle la raiz
    s = s**(1/2)
    s= round(s,3)
    return s
    
def moda(lista):
    usados = []
    modas = []
    for i in lista:
        if i not in usados:
            n = lista.count(i)
            modas.append(n)
            usados.append(i)
    maximo = max(modas) #el valor mas repetido
    
    #vemos cuantos tienen esta frecuencia
    cuantos = 0
    modaValor = []
    for f in range(len(modas)):
        if modas[f] == maximo:
            cuantos = cuantos + 1
            modaValor.append(usados[f])
    
    if cuantos == len(modas):
        respuesta = "No hay moda"
    
    else:
        respuesta = modaValor
    return respuesta

            
#agregamos n
n = validar(1,100,"Ingrese un valor mayor a 1 y menor o igual a 100: ")

#agregamos n valores
valores = []
for i in range(n):
    v = int(input("Ingrese un valor ="))
    valores.append(v)

me = media(valores)
d = desv(me,valores)
mo = moda(valores)

print(me)
print(d)
print(mo)
```

```python
## 3.
names = ["Matias", "Marcela", "Felipe", "Alberto", "Fabiola", "Margarita", "Benjamin"]

sorted_names = ["Matias", "Marcela", "Felipe", "Alberto", "Fabiola", "Margarita", "Benjamin"]
sorted_names.sort()

min_name = sorted_names[0]

output = []
output.append(min_name)

for name in names:
  if name != min_name:
    output.append(name)

print(output)
```

```python
## 4.
grades = []

while True:
  grade = float(input("> "))
  while not (0 <= grade <= 7):
    grade = float(input("> "))
  
  if grade == 0: break
  grades.append(grade)

counters = [0 for i in range(6)]
for grade in grades:
  if (1 <= grade < 2):
    counters[0] += 1
  elif (2 <= grade < 3):
    counters[1] += 1
  elif (3 <= grade < 4):
    counters[2] += 1
  elif (4 <= grade < 5):
    counters[3] += 1
  elif (5 <= grade < 6):
    counters[4] += 1
  elif (6 <= grade < 7):
    counters[5] += 1

print("Histograma de notas: ")
for interval in range(len(counters)):
  if interval == 5:
    print(f"{interval + 1}.0 - {interval + 2}.0 | {'*' * counters[interval]}")
  else:
    print(f"{interval + 1}.0 - {interval + 2}.9 | {'*' * counters[interval]}")
```

```python
# 5. 
def validar(min, max, msg):
  v = int(input(msg))
  while v <= min or v >= max:
    print("Error, ingrese valor dentro del rango.")
    v = int(input(msg))
  return v

def AgrProductos(mensaje, lista_super):
  prod = input(mensaje)
  if prod not in lista_super: # Verifica si acaso el producto no se encuentra en la lista
    lista_super.append(prod) # Agrega el producto a la lista
  
  return lista_super

def main():
  N = validar(1, 100, "Ingrese la cantidad de productos: ")
  
  # Creamos la lista de supermercado desde main, para que sea accesible desde todas las otras funciones
  lista_super = []
  for i in range(N):
    AgrProductos("> Ingrese el nombre del producto: ", lista_super)
  
  # Imprimimos cada item de la lista, pero ahora ordenada
  lista_super.sort()
  for i in lista_super:
    if i == lista_super[-1]: # Si el item, es el Ãºltimo de la lista
      print(i) # Lo imprimimos sin una coma de separacion
    else:
      print(i, end=", ") # Imprimimos con coma de separacion      

main()
```
