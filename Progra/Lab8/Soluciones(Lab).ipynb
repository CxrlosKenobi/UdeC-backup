```python
# 1.- La Casa de Papel
def validate_input(minimo, maximo):
  answer = int(input("> "))

  while not (minimo < answer < maximo):
    print(f"Input fuera de rango ({minimo}, {maximo})")
    answer = int(input("> "))
  
  return answer

def only_4s_and_7s(n):
  while n > 0:
    if ( ((n % 10) != 4) and ((n % 10) != 7) ):
      return False

    n = n // 10
  
  return True

def main():
  input_lines = validate_input(1, 1000)

  for i in range(input_lines):
    n = validate_input(1, 1000)
    
    if (
      (n % 4 == 0) or (n % 7 == 0) or (only_4s_and_7s(n))
    ):
      print("Número especial")
    else:
      print("Número normal")

main()
```

```python
# 2.- Multiplicación Egipcia (Cami)
def validar(vmenor, vmayor, mensaje):
  v = int(input(mensaje))
  while v<vmenor or v>vmayor:
    print("Valor fuera de rango, debe estar en el rango(" + str(vmenor) + "," + str(vmayor) + ")")
    v = int(input(mensaje))
  return v
def signo(numero):
  if numero <0:
    return "Negativo"
  else:
    return "Positivo"
def operacion(op1,op2):
  resultado = 0
  while True:
    #hay que ver si el op1 es impar
    if op1 % 2 != 0:
      resultado = resultado + op2
    
    #ahora calculamos la mitad de op1 y el doble de op2
    mitad = op1 // 2
    doble = op2 * 2

    #actualizamos los datos
    op1 = mitad
    op2 = doble

    if op1<1:
      break #cortamos el ciclo
    
  
  return resultado

#PROGRAMA PRINCIPAL
op1 = validar(1,float("inf"), "Ingrese el operando 1: ")
op2 = validar(1,float("inf"), "Ingrese el operando 2: ")

result = operacion(op1,op2)
print("El resultado es: " + str(result))
```

```python
# 2.- Multiplicación Egipcia
def validate_input(minimo, maximo):
  answer = int(input("> "))

  while not (minimo < answer < maximo):
    print(f"Input fuera de rango ({minimo}, {maximo})")
    answer = int(input("> "))
  
  return answer

def signo(numero):
  if numero <0:
    return "Negativo"
  else:
    return "Positivo"
def mitades_y_dobles(A, B):
  As = [A]; Bs = [B]
  print(f"\n{A}\t{B}")

  while A != 1:
    A //= 2; B *= 2
    print(f"{A}\t{B}")
    As.append(A); Bs.append(B)

  return As, Bs

def eliminaciones(As, Bs):
  to_remove = []
  for i in range(len(Bs)):
    if (As[i] % 2 == 0):
      print(f"Se elimina el {Bs[i]} por estar delante de un {As[i]} (par)")
      to_remove.append(Bs[i])
    else:
      print(f"Se mantiene el {Bs[i]}")

  for i in range(len(to_remove)):
    Bs.remove(to_remove[i])

  return Bs

def main():
  print("Ingrese A: ", end="")
  A = validate_input(1, 10000); print(A)
  print("Ingrese B: ", end="")
  B = validate_input(1, 10000); print(B)

  As, Bs = mitades_y_dobles(A, B)
  Bs = eliminaciones(As, Bs)

  print("Resultado: ", sum(Bs))

main()
```

```python
# 3.- Sistema de votaciones CC

# - Leer la cantidad de votos
# - Leer y contabilizar los votos de cada categoria
# - Desplegar los resultados

def validar_input(minimo, maximo):
  answer = int(input("> "))

  while not (minimo < answer < maximo):
    print(f"Input fuera de rango ({minimo}, {maximo})")
    answer = int(input("> "))
  
  return answer

def leer_voto():
  voto = input('Ingrese su voto (F, C, X o _: ')

  while (voto != 'F') and (voto != 'C') and (voto != 'X') and (voto != '_'):
    print('Error, voto mal ingresado, intente nuevamente.')
    voto = input('Ingrese su voto (F, C, X o _: ')

  return voto

def obtener_resultados(cant_votantes):
  f, c, x, nulo = 0, 0, 0, 0

  for i in range(cant_votantes):
    voto = leer_voto()
    if voto == 'F':
      f += 1
    elif voto == 'C':
      c += 1
    elif voto == 'X':
      x += 1
    else:
      nulo += 1

  return f, c, x, nulo

def desplegar_resultados(f, c, x, nulo):
  print(f"\nF: {f}\tC: {c}\tX: {x}\tNulo: {nulo}\n")
  
  if ((f + nulo) / (f + c + nulo)) > 2/3:
    print("[ ok ] Iniciativa aprobada")
  else:
    print("[ ! ] Iniciativa rechazada")


def main():
  cant_votantes = validar_input(1, 1000)
  f, c, x, nulo = obtener_resultados(cant_votantes)
  desplegar_resultados(f, c, x, nulo)

main()
```

```python
# 4.- Prodigio Matemático
def validar_input(minimo, maximo):
  answer = int(input("> "))

  while not (minimo < answer < maximo):
    print(f"Input fuera de rango ({minimo}, {maximo})")
    answer = int(input("> "))
  
  return answer

def factorial(n):
  f = 1
  while n > 0:
    f = f * n
    n = n - 1
  
  return f

def zero_detector(f):
  zeros, not_zeros = 0, 0

  while f > 0:
    d = f % 10
    f //= 10

    if d == 0:
      zeros += 1
    else:
      not_zeros += 1
  
  return zeros, not_zeros

def main():
  n = validar_input(1, 100)
  f = factorial(n)
  zeros, not_zeros = zero_detector(f)

  print(zeros, not_zeros)

main()
```
